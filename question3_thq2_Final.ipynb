```python
grades =[]

for row in dataset['Total.marks']:
    if row > 84:
        grades.append('HD')
    elif row > 74:
        grades.append('D')
    elif row > 64:
        grades.append('Cr')
    elif row > 49:
        grades.append('P')
    else:
        grades.append('F')
```

```python
dataset['Grade'] = grades
```
